# DATA VISUALIZATION AND EXPLORATION

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#import dataset
from sklearn.datasets import load_iris
```

```python
iris_data=load_iris()
```

```python

df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df.head()
```

```python
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
```

# 1. Comparison between various species based on sepal length and width using respective visualization

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'],c=iris_data.target)
```

```python
import seaborn as sns
sns.scatterplot(df['sepal length (cm)'],df['sepal width (cm)'],hue =df['species'],s=45)
```

```python
sns.scatterplot(df['petal length (cm)'],df['petal width (cm)'],hue =df['species'],s=45)
```

# 2. Perform the Bi-variate Analysis via pairplot visualization

```python
sns.pairplot(df,hue="species")
```

# 3. Justify Sepal Length and Sepal Width features are slightly correlated with each other

```python
sns.heatmap(df.corr(),annot=True)
```

# 4. Visualize the categorical feature “Species” with all other four input variables

```python
fig, axes = plt.subplots(2, 2, figsize=(14,9))
sns.boxplot( y='petal length (cm)', x= 'species', data=df , ax=axes[0, 0])
sns.boxplot( y='sepal length (cm)', x= 'species', data=df , ax=axes[0, 1])
sns.boxplot( y='sepal width (cm)', x= 'species', data=df , ax=axes[1, 0])
sns.boxplot( y='petal width (cm)', x= 'species', data=df, ax=axes[1, 1])
plt.show()
```

# 5. Plot the density of the length and width in the species.

```python
fig, axes = plt.subplots(2, 2, figsize=(14,9))
sns.violinplot( y='petal length (cm)', x= 'species', data=df , ax=axes[0, 0])
sns.violinplot( y='sepal length (cm)', x= 'species', data=df , ax=axes[0, 1])
sns.violinplot( y='sepal width (cm)', x= 'species', data=df , ax=axes[1, 0])
sns.violinplot( y='petal width (cm)', x= 'species', data=df, ax=axes[1, 1])
plt.show()
```
